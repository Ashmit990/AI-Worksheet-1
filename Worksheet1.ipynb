Problem - 4: Matrix Operations:
• For the following arrays:
A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);
Prove following with Numpy:
1. Prove A.A−1 = I.

```python
A = np.array([[3, 4],[7, 8]])
A_inv = np.linalg.inv(A)
I = np.dot(A, A_inv)
print(I)
```

2. Prove AB ̸= BA.

```python
import numpy as np

A = np.array([[3, 4],[7, 8]])
B = np.array([[5, 3],[2, 1]])

AB = np.dot(A, B)
BA = np.dot(B, A)

print("AB=", AB)
print("BA=", BA)
```

3. Prove (AB)

T = BTAT

```python
import numpy as np

A = np.array([[3, 4],[7, 8]])
B = np.array([[5, 3],[2, 1]])

left = np.dot(A, B).T
right = np.dot(B.T, A.T)

print("left =", left)
print("right =", right)
print("equal =", np.array_equal(left, right))
```
